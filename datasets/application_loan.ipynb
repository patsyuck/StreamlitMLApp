```python
df = data[['TARGET', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'CNT_CHILDREN', 
           'AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE', 'NAME_INCOME_TYPE', 
           'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE', 'DAYS_BIRTH']]
df.tail()
```

```python
df = df.rename(columns={'TARGET': 'target', 'CODE_GENDER': 'gender', 'FLAG_OWN_CAR': 'hasCar', 'FLAG_OWN_REALTY': 'hasRealty', 
        'CNT_CHILDREN': 'cntChildren', 'AMT_INCOME_TOTAL': 'income', 'AMT_CREDIT': 'credit', 'AMT_ANNUITY': 'annuity', 
        'AMT_GOODS_PRICE': 'goodsPrice', 'NAME_INCOME_TYPE': 'incomeType', 'NAME_EDUCATION_TYPE': 'education', 
        'NAME_FAMILY_STATUS': 'family', 'NAME_HOUSING_TYPE': 'apartment', 'DAYS_BIRTH': 'birth'})
df.tail()
```

```python
df.loc[:, 'gender'] = df['gender'].apply(lambda x: 0 if x == 'F' else (1 if x == 'M' else np.nan))
df.loc[:, 'hasCar'] = df['hasCar'].apply(lambda x: 0 if x == 'N' else 1)
df.loc[:, 'hasRealty'] = df['hasRealty'].apply(lambda x: 0 if x == 'N' else 1)
df.loc[:, 'cntChildren'] = df['cntChildren'].apply(lambda x: x if x < 6 else 6)
df.loc[:, 'income'] = df['income'].apply(lambda x: x // 50000 if x < 500000 else 10)
df.loc[:, 'credit'] = df['credit'].apply(lambda x: x // 50000 if x < 500000 else 10)
df.loc[:, 'annuity'] = df['annuity'].apply(lambda x: x // 10000 if x < 100000 else 10)
df.loc[:, 'goodsPrice'] = df['goodsPrice'].apply(lambda x: x // 200000 if x < 2000000 else 10)
df.loc[:, 'IsJobWorking'] = df['incomeType'].apply(lambda x: 1 if x == 'Working' else 0)
df.loc[:, 'IsJobCommercial'] = df['incomeType'].apply(lambda x: 1 if x == 'Commercial associate' else 0)
df.loc[:, 'IsJobState'] = df['incomeType'].apply(lambda x: 1 if x == 'State servant' else 0)
df.loc[:, 'IsJobPensioner'] = df['incomeType'].apply(lambda x: 1 if x == 'Pensioner' else 0)
df.loc[:, 'IsEducLowerSec'] = df['education'].apply(lambda x: 1 if x == 'Lower secondary' else 0)
df.loc[:, 'IsEducSecondary'] = df['education'].apply(lambda x: 1 if x == 'Secondary / secondary special' else 0)
df.loc[:, 'IsEducHigherInc'] = df['education'].apply(lambda x: 1 if x == 'Incomplete higher' else 0)
df.loc[:, 'IsEducHigher'] = df['education'].apply(lambda x: 1 if x == 'Higher education' else 0)
df.loc[:, 'IsFamMarried'] = df['family'].apply(lambda x: 1 if x == 'Married' else 0)
df.loc[:, 'IsFamNotMarried'] = df['family'].apply(lambda x: 1 if x == 'Single / not married' else 0)
df.loc[:, 'IsFamCivil'] = df['family'].apply(lambda x: 1 if x == 'Civil marriage' else 0)
df.loc[:, 'IsFamSeparated'] = df['family'].apply(lambda x: 1 if x == 'Separated' else 0)
df.loc[:, 'IsFamWidow'] = df['family'].apply(lambda x: 1 if x == 'Widow' else 0)
df.loc[:, 'IsApartHouse'] = df['apartment'].apply(lambda x: 1 if x == 'House / apartment' else 0)
df.loc[:, 'IsApartParents'] = df['apartment'].apply(lambda x: 1 if x == 'With parents' else 0)
df.loc[:, 'IsApartMunicipal'] = df['apartment'].apply(lambda x: 1 if x == 'Municipal apartment' else 0)
df.loc[:, 'IsApartRented'] = df['apartment'].apply(lambda x: 1 if x == 'Rented apartment' else 0)
df.loc[:, 'IsApartOffice'] = df['apartment'].apply(lambda x: 1 if x == 'Office apartment' else 0)
df.loc[:, 'age'] = df['birth'].apply(lambda x: abs(x) // 365)
df.tail()
```

```python
df = df.drop(['incomeType', 'education', 'family', 'apartment', 'birth'], axis=1)
df.tail()
```

```python
df.to_csv('loan_train.csv', index=False)
```
